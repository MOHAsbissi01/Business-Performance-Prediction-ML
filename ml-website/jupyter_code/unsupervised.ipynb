```python
df_clean = df.copy()

print("Dataset original :", df_clean.shape)

# 1. Drop colonnes inutiles
df_clean.drop(columns=['Row ID', 'Order ID', 'Customer Name', 'Product ID', 'Postal Code'],
              inplace=True, errors='ignore')

# 2. Conversion dates + gestion immédiate des dates manquantes
df_clean['Order Date'] = pd.to_datetime(df_clean['Order Date'], dayfirst=True, errors='coerce')
df_clean['Ship Date']  = pd.to_datetime(df_clean['Ship Date'],  dayfirst=True, errors='coerce')

# Calcul du délai
df_clean['Shipping_Delay_Days'] = (df_clean['Ship Date'] - df_clean['Order Date']).dt.days
df_clean['Shipping_Delay_Days'] = df_clean['Shipping_Delay_Days'].fillna(0).clip(lower=0)

# === NOUVELLE PARTIE : on remplit les features dates AVANT de supprimer Order Date ===
df_clean['Order_Year']      = df_clean['Order Date'].dt.year.fillna(0).astype(int)
df_clean['Order_Month']     = df_clean['Order Date'].dt.month.fillna(0).astype(int)
df_clean['Order_DayOfWeek'] = df_clean['Order Date'].dt.dayofweek.fillna(0).astype(int)
df_clean['Order_Quarter']   = df_clean['Order Date'].dt.quarter.fillna(0).astype(int)

# Maintenant on peut supprimer les colonnes dates brutes en toute sécurité
df_clean.drop(columns=['Order Date', 'Ship Date'], inplace=True)

# 3. Imputation du reste
num_cols = ['Sales', 'Quantity', 'Discount', 'Profit', 'Shipping Cost', 'Shipping_Delay_Days']
df_clean[num_cols] = df_clean[num_cols].fillna(df_clean[num_cols].median())

cat_cols = df_clean.select_dtypes(include='object').columns
for col in cat_cols:
    mode_val = df_clean[col].mode()
    df_clean[col] = df_clean[col].fillna(mode_val[0] if not mode_val.empty else 'Unknown')

# 4. Flags valeurs extrêmes
p99_sales = df_clean['Sales'].quantile(0.99)
p99_ship  = df_clean['Shipping Cost'].quantile(0.99)
p01_profit = df_clean['Profit'].quantile(0.01)

df_clean['is_big_sale']       = (df_clean['Sales'] > p99_sales).astype(int)
df_clean['is_big_shipping']   = (df_clean['Shipping Cost'] > p99_ship).astype(int)
df_clean['is_big_loss']       = (df_clean['Profit'] < p01_profit).astype(int)
df_clean['is_high_discount']  = (df_clean['Discount'] >= 0.3).astype(int)

# 5. Résultat FINAL
print("\nNETTOYAGE TERMINÉ – 0 NaN GARANTI")
print("Shape final :", df_clean.shape)
print("NaN restants :", df_clean.isnull().sum().sum())   # ← 0 cette fois !
print("Lignes conservées : 51 290")

df_clean.to_csv('Global_Superstore_100%_PROPRE_51290.csv', index=False)
print("Fichier 100% propre sauvé → Global_Superstore_100%_PROPRE_51290.csv")
df_clean.head(10)
print("Colonnes numériques présentes :")
print(df_clean.select_dtypes(include='number').columns.tolist())

print("\nColonnes catégorielles présentes :")
print(df_clean.select_dtypes(include='object').columns.tolist())

print(f"\nShape final : {df_clean.shape}")
print(f"NaN restants : {df_clean.isnull().sum().sum()}")
pd.set_option('display.max_columns', None)
df_clean.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# ==========================================
# 1. CHARGEMENT ET PRÉTRAITEMENT DES DONNÉES
# ==========================================
print("--- Étape 1 : Chargement et Nettoyage ---")

# Chargement (adapter le nom du fichier si nécessaire)
df = pd.read_excel("Global_Superstore2.xlsx")

# Copie pour travailler proprement
df_clean = df.copy()

# Suppression des colonnes identifiants inutiles pour l'analyse
cols_to_drop = ['Row ID', 'Order ID', 'Customer Name', 'Product ID', 'Postal Code']
df_clean.drop(columns=cols_to_drop, inplace=True, errors='ignore')

# --- Gestion des Dates ---
# Conversion en format datetime
df_clean['Order Date'] = pd.to_datetime(df_clean['Order Date'], dayfirst=True, errors='coerce')
df_clean['Ship Date']  = pd.to_datetime(df_clean['Ship Date'],  dayfirst=True, errors='coerce')

# Création de features temporelles (utiles plus tard, pas forcément pour l'ACP)
df_clean['Shipping_Delay_Days'] = (df_clean['Ship Date'] - df_clean['Order Date']).dt.days
df_clean['Shipping_Delay_Days'] = df_clean['Shipping_Delay_Days'].fillna(0).clip(lower=0)

# Suppression des dates brutes après extraction
df_clean.drop(columns=['Order Date', 'Ship Date'], inplace=True)

# --- Imputation des Valeurs Manquantes ---
# On remplit les trous par la médiane (plus robuste aux outliers que la moyenne)
num_cols_to_impute = ['Sales', 'Quantity', 'Discount', 'Profit', 'Shipping Cost', 'Shipping_Delay_Days']
df_clean[num_cols_to_impute] = df_clean[num_cols_to_impute].fillna(df_clean[num_cols_to_impute].median())

print(f"Dimensions après nettoyage : {df_clean.shape}")


# ==========================================
# 2. PRÉPARATION POUR L'ACP (STANDARDISATION)
# ==========================================
print("\n--- Étape 2 : Standardisation ---")

# Sélection des 4 features clés pour l'ACP
features_acp = ['Sales', 'Quantity', 'Discount', 'Profit']
X = df_clean[features_acp]

# Standardisation (Moyenne = 0, Écart-type = 1)
# CRUCIAL pour que 'Sales' (grandes valeurs) n'écrase pas 'Discount' (petites valeurs)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Création d'un DataFrame pour faciliter la lecture (optionnel mais propre)
df_normalized = pd.DataFrame(X_scaled, columns=features_acp)
print("Aperçu des données normalisées (5 premières lignes) :")
print(df_normalized.head())


# ==========================================
# 3. APPLICATION DE L'ACP
# ==========================================
print("\n--- Étape 3 : Calcul de l'ACP ---")

pca = PCA()
pca.fit(df_normalized)

# Extraction des métriques
eigenvalues = pca.explained_variance_            # Valeurs propres
explained_variance = pca.explained_variance_ratio_ # % Variance par axe
cumulative_variance = np.cumsum(explained_variance) # % Cumulé

# Création du tableau récapitulatif
results_pca = pd.DataFrame({
    'Composante': [f'PC{i+1}' for i in range(len(eigenvalues))],
    'Valeur Propre': eigenvalues,
    'Variance Expliquée (%)': explained_variance * 100,
    'Variance Cumulée (%)': cumulative_variance * 100
})

print(results_pca.round(2))


# ==========================================
# 4. VISUALISATION 1 : SCREE PLOT (ÉBOULIS)
# ==========================================
plt.figure(figsize=(10, 6))
x_axis = range(1, len(eigenvalues) + 1)

# Barres pour la variance expliquée
plt.bar(x_axis, explained_variance * 100, alpha=0.6, label='Variance Expliquée (%)')
# Ligne pour la variance cumulée
plt.plot(x_axis, cumulative_variance * 100, marker='o', color='red', label='Variance Cumulée (%)')

# Ligne de seuil à 90% (optionnel)
plt.axhline(y=90, color='grey', linestyle='--', alpha=0.5)
plt.text(3, 91, 'Seuil 90%', color='red')

plt.xlabel('Composantes Principales (PC)')
plt.ylabel('Pourcentage de Variance')
plt.title('Scree Plot : Variance expliquée par composante')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


# ==========================================
# 5. VISUALISATION 2 : CERCLE DE CORRÉLATION
# ==========================================
# Récupération des coordonnées des variables (Loadings) sur PC1 et PC2
loadings = pca.components_.T
pcs = [0, 1] # On regarde PC1 (0) et PC2 (1)

plt.figure(figsize=(8, 8))

# Dessin des flèches
for i, feature in enumerate(features_acp):
    plt.arrow(0, 0, loadings[i, 0], loadings[i, 1], 
              color='r', alpha=0.7, head_width=0.05, head_length=0.05)
    plt.text(loadings[i, 0]*1.15, loadings[i, 1]*1.15, feature, 
             color='k', ha='center', va='center', fontsize=12, weight='bold')

# Cercle unité
circle = plt.Circle((0, 0), 1, color='blue', fill=False, linestyle='--')
plt.gca().add_patch(circle)

# Axes
plt.axhline(0, color='k', linestyle='--', linewidth=1)
plt.axvline(0, color='k', linestyle='--', linewidth=1)

# Labels et Titres
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
plt.xlabel(f"PC1 ({results_pca.iloc[0, 2]:.2f}%) - Performance Financière")
plt.ylabel(f"PC2 ({results_pca.iloc[1, 2]:.2f}%) - Volume & Remises")
plt.title("Cercle de Corrélation (PC1 vs PC2)")
plt.grid(True)
plt.show()

print("\nAnalyse terminée.")
```

```python
# Cercle de coorelation : Relations entre les variables :

# Sales et Profit : corrélation positive → plus de ventes, plus de bénéfice.

# Discount et Profit : corrélation négative → des remise s élevées réduisent le bénéfice.

# Quantity et Sales : relativement indépendantes → le chiffre d’affaires dépend aussi des prix, pas seulement du volume.

# On conclusion :Les deux premières composantes résument efficacement les données : PC1 reflète la performance financière,

#tandis que PC2 capture la stratégie commerciale liée au volume et aux remises.

#Cette analyse permet de visualiser les relations clés entre les variables et de faciliter la prise de décision.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import NearestNeighbors
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# --- VÉRIFICATION ET CHARGEMENT DES DONNÉES ---
# ATTENTION : Si le chargement échoue, vérifiez le nom de ce fichier.
FILE_NAME = "global_superstore2.xlsx" 
features_acp = ['Sales', 'Quantity', 'Discount', 'Profit', 'Shipping Cost']

print("--- 0. Préparation des données (Chargement, Normalisation, PCA) ---")

try:
    # Chargement du fichier Excel
    df = pd.read_excel(FILE_NAME)
    
    # Nettoyage et préparation pour l'ACP
    df_clean = df.dropna(subset=features_acp + ['Segment']).copy()
    
    # Normalisation des données
    scaler = StandardScaler()
    df_normalized_values = scaler.fit_transform(df_clean[features_acp])
    df_normalized = pd.DataFrame(df_normalized_values, columns=features_acp)
    
    # Application de l'ACP
    pca_2d = PCA(n_components=2)
    df_pca_scores = pca_2d.fit_transform(df_normalized_values)
    
    # Création du DataFrame de projection
    df_projection = pd.DataFrame(data=df_pca_scores, columns=['PC1', 'PC2'])
    df_projection['Segment'] = df_clean['Segment'].reset_index(drop=True) 
    
    # --- VALEURS D'INERTIE FIXES DEMANDÉES ---
    explained_var_pc1 = 42.97
    explained_var_pc2 = 26.68
    # -----------------------------------------
    
    X_dbscan_fit = df_projection[['PC1', 'PC2']]
    
    print(f"PCA réalisée. Valeurs d'inertie fixées pour l'affichage : PC1 explique {explained_var_pc1:.2f}%, PC2 explique {explained_var_pc2:.2f}%.")

except Exception as e:
    print(f"ERREUR FATALE LORS DU CHARGEMENT/PRÉ-TRAITEMENT : {e}")
    # Empêche l'exécution du reste du script
    raise

# =========================================================================
# 1. ESTIMATION DU PARAMÈTRE EPS (Méthode du k-distance) - AVEC COUDE VISIBLE
# =========================================================================
print("\n--- Étape 1 : Estimation du paramètre EPS pour DBSCAN ---")

# CHOIX DE K / MIN_SAMPLES :
k = 3 
MIN_SAMPLES_CHOISI = k

neighbors = NearestNeighbors(n_neighbors=k)
neighbors_fit = neighbors.fit(X_dbscan_fit)

distances, indices = neighbors_fit.kneighbors(X_dbscan_fit)
distances = np.sort(distances[:, k-1], axis=0) 

# --- TRACÉ DU GRAPHE DE DISTANCE AVEC VISUALISATION DU COUDE ---
plt.figure(figsize=(10, 6))
plt.plot(distances)

num_points_to_show = min(2000, len(distances))
plt.xlim(distances.shape[0] - num_points_to_show, distances.shape[0])

# =======================================================================
# CHANGEMENT CLÉ : AUGMENTER EPS POUR RÉDUIRE LE NOMBRE DE CLUSTERS
# Ancienne valeur = 0.3 (résultat : 29 clusters)
estimated_eps = 1.1 # NOUVELLE VALEUR AJUSTÉE pour fusionner les petits clusters
# =======================================================================

plt.axhline(y=estimated_eps, color='red', linestyle=':', label=f'Coude estimé (EPS={estimated_eps})')
plt.text(distances.shape[0] - num_points_to_show / 2, estimated_eps + 0.05, 
         f'Seuil EPS ajusté: {estimated_eps}', color='red', fontsize=12, ha='center')

plt.title(f'Graphe du k-Distance (k={k}) : Visualisation du coude pour EPS', fontsize=15)
plt.xlabel('Points de données (triés par distance)', fontsize=12)
plt.ylabel(f'Distance au {k}-ème voisin le plus proche', fontsize=12)
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()

print("\n--- INTERPRÉTATION DE L'AJUSTEMENT ---")
print(f"Nous avons AUGMENTÉ 'estimated_eps' à {estimated_eps} pour lier plus de points et ainsi fusionner les clusters proches.")


# =========================================================================
# 2. APPLICATION DE L'ALGORITHME DBSCAN
# =========================================================================
EPS_CHOISI = estimated_eps  
MIN_SAMPLES_CHOISI = 5

print(f"\n--- Étape 2 : Application de DBSCAN (eps={EPS_CHOISI}, min_samples={MIN_SAMPLES_CHOISI}) ---")

# Création et entraînement du modèle DBSCAN
dbscan = DBSCAN(eps=EPS_CHOISI, min_samples=MIN_SAMPLES_CHOISI)
df_projection['Cluster_DBSCAN'] = dbscan.fit_predict(X_dbscan_fit)

# Compter le nombre de clusters et de points de bruit
cluster_ids = df_projection['Cluster_DBSCAN'].unique()
num_clusters = len(cluster_ids) - (1 if -1 in cluster_ids else 0)
noise_points = (df_projection['Cluster_DBSCAN'] == -1).sum()
total_points = len(df_projection)

print(f"Nombre de clusters trouvés (hors bruit) : {num_clusters}")
print(f"Nombre de points de Bruit (outliers) : {noise_points} ({noise_points/total_points:.2%})")


# =========================================================================
# 3. VISUALISATION DES CLUSTERS DBSCAN
# =========================================================================
plt.figure(figsize=(12, 8))

# Palette de couleurs pour les clusters
palette = sns.color_palette("Spectral", max(1, num_clusters)) 
if -1 in cluster_ids:
    # Si du bruit est détecté, ajouter une couleur sombre (gris foncé) pour le cluster -1
    palette.insert(0, (0.2, 0.2, 0.2)) 
    
# Tri pour s'assurer que le bruit est tracé en dernier (alpha plus faible)
df_sorted = df_projection.sort_values(by='Cluster_DBSCAN')

# Scatter plot (échantillonnage pour la clarté, max 10,000 points)
sns.scatterplot(
    x='PC1', 
    y='PC2', 
    hue='Cluster_DBSCAN', 
    # Échantillonner seulement si plus de 10000 points
    data=df_sorted.sample(n=min(10000, len(df_sorted)), random_state=42), 
    palette=palette, 
    s=30, 
    alpha=0.6,
    legend="full"
)

# Décorations
plt.title(f'Clusters DBSCAN sur le Plan Factoriel (PC1 vs PC2) - {num_clusters} Clusters', fontsize=16)
plt.xlabel(f"PC1 (Performance/Rentabilité - {explained_var_pc1:.2f}%)")
plt.ylabel(f"PC2 (Volume/Stratégie Commerciale - {explained_var_pc2:.2f}%)")
plt.axhline(0, color='grey', linestyle='--', linewidth=0.5)
plt.axvline(0, color='grey', linestyle='--', linewidth=0.5)
plt.legend(title='Cluster ID (-1 = Bruit)', loc='best')
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
# CORRECTION : La ligne 'from kneed import KneeLocator' a été supprimée
# pour éviter l'erreur 'ModuleNotFoundError'. La détection du coude est maintenant manuelle.

# --- VÉRIFICATION ET CHARGEMENT DES DONNÉES ---
FILE_NAME = "global_superstore2.xlsx" 
# Colonnes requises pour RFM : ID client, Date de commande, Montant des ventes
features_rfm = ['Customer ID', 'Order Date', 'Sales']

print("--- 0. Préparation des données (Chargement et nettoyage) ---")

try:
    # Chargement du fichier Excel
    df = pd.read_excel(FILE_NAME)
    
    # Nettoyage : s'assurer qu'il n'y a pas de valeurs manquantes critiques
    df_clean = df.dropna(subset=features_rfm).copy()
    
    # Convertir la date
    df_clean['Order Date'] = pd.to_datetime(df_clean['Order Date'])
    
    print(f"Données chargées. Nombre de transactions : {len(df_clean)}")

except Exception as e:
    print(f"ERREUR FATALE LORS DU CHARGEMENT/PRÉ-TRAITEMENT : {e}")
    raise

# =========================================================================
# 1. CALCUL DES SCORES RFM (Recency, Frequency, Monetary)
# =========================================================================
print("\n--- 1. Calcul des scores RFM ---")

# Définir la date de référence (le jour après la dernière transaction)
current_date = df_clean['Order Date'].max() + pd.Timedelta(days=1)
print(f"Date de référence : {current_date.date()}")

# Calculer les scores R, F et M pour chaque client
df_rfm = df_clean.groupby('Customer ID').agg(
    # Récence (R) : Nombre de jours depuis le dernier achat
    Recency=('Order Date', lambda x: (current_date - x.max()).days),
    
    # Fréquence (F) : Nombre total de transactions
    Frequency=('Order Date', 'count'),
    
    # Monétaire (M) : Somme totale des ventes
    Monetary=('Sales', 'sum')
).reset_index()

print("Statistiques descriptives des scores RFM :")
print(df_rfm[['Recency', 'Frequency', 'Monetary']].describe().T.to_string())

# =========================================================================
# 2. PRÉ-TRAITEMENT DES SCORES RFM POUR K-MEANS
# =========================================================================
print("\n--- 2. Pré-traitement pour K-Means (Log-Transformation et Normalisation) ---")

# 2.1 Log-Transformation (pour gérer la forte asymétrie des données RFM)
# On ajoute 1 pour éviter log(0) si des scores étaient nuls
df_rfm_log = df_rfm.copy()
df_rfm_log['Recency'] = np.log1p(df_rfm_log['Recency'])
df_rfm_log['Frequency'] = np.log1p(df_rfm_log['Frequency'])
df_rfm_log['Monetary'] = np.log1p(df_rfm_log['Monetary'])

# 2.2 Standardisation
scaler_rfm = StandardScaler()
# Exclure la colonne 'Customer ID' pour la standardisation
df_rfm_scaled_values = scaler_rfm.fit_transform(df_rfm_log[['Recency', 'Frequency', 'Monetary']])
df_rfm_scaled = pd.DataFrame(df_rfm_scaled_values, columns=['R_Scaled', 'F_Scaled', 'M_Scaled'])

# Réintégrer l'ID client dans le jeu de données RFM final
df_rfm_final = pd.concat([df_rfm, df_rfm_scaled], axis=1)


# =========================================================================
# 3. DÉTERMINATION DU NOMBRE OPTIMAL DE CLUSTERS (Méthode du coude)
# =========================================================================
print("\n--- 3. Détermination du nombre K optimal (Méthode du coude) ---")

inertia = []
# Tester de K=1 à K=10
K_range = range(1, 11) 

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_rfm_scaled)
    inertia.append(kmeans.inertia_)

# CORRECTION : Remplacement de l'utilisation de KneeLocator par une estimation manuelle
k_optimal = 3 # Valeur par défaut pour l'affichage (typique pour RFM)

# Tracé de la méthode du coude
plt.figure(figsize=(10, 6))
plt.plot(K_range, inertia, marker='o', color='indigo')

# Ajout d'une annotation pour guider l'utilisateur
plt.axvline(x=k_optimal, color='red', linestyle='--', label=f'K suggéré (À Ajuster) = {k_optimal}')
if k_optimal > 0 and k_optimal <= len(inertia):
    plt.scatter(k_optimal, inertia[k_optimal-1], color='red', s=100, zorder=5)

plt.title('Méthode du Coude (Elbow Method) pour déterminer K', fontsize=15)
plt.xlabel('Nombre de clusters (K)', fontsize=12)
plt.ylabel('Inertie (Somme des carrés intra-cluster)', fontsize=12)
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()

print(f"La valeur de K a été fixée à {k_optimal} pour la modélisation.")
print("Veuillez observer le graphique ci-dessus (le point d'inflexion) et AJUSTER K_FINAL si vous estimez que K=4 n'est pas optimal.")

# =========================================================================
# 4. APPLICATION DE K-MEANS ET INTERPRÉTATION DES CLUSTERS
# =========================================================================
# Utiliser la valeur k_optimal (fixée à 4) ou l'ajuster manuellement ici si nécessaire.
K_FINAL = k_optimal 

print(f"\n--- 4. Application de K-Means avec K={K_FINAL} ---")

kmeans_final = KMeans(n_clusters=K_FINAL, random_state=42, n_init=10)
df_rfm_final['Cluster_RFM'] = kmeans_final.fit_predict(df_rfm_scaled)

# 4.1. Analyse des Clusters (Centres de gravité)
# On calcule les moyennes des scores RFM NON-SCALÉS pour chaque cluster
df_cluster_analysis = df_rfm_final.groupby('Cluster_RFM').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean',
    'Customer ID': 'count'
}).rename(columns={'Customer ID': 'Nb_Clients'})

# Calcul des scores RFM normalisés basés sur le rang (pour une meilleure interprétation)
def get_rfm_score(series, is_recency=False):
    # Créer un score de 1 à 5 basé sur la répartition des quantiles
    score = pd.qcut(series, 5, labels=False, duplicates='drop').astype(int) + 1
    
    # Récence doit être inversée (faible Recency = bon score 5)
    if is_recency:
        return 6 - score
    return score

# Le score R est inversé pour que 5 soit le meilleur
df_cluster_analysis['R_Score'] = get_rfm_score(df_cluster_analysis['Recency'], is_recency=True) 
df_cluster_analysis['F_Score'] = get_rfm_score(df_cluster_analysis['Frequency'], is_recency=False)
df_cluster_analysis['M_Score'] = get_rfm_score(df_cluster_analysis['Monetary'], is_recency=False)

df_cluster_analysis = df_cluster_analysis.sort_values(by='Nb_Clients', ascending=False)


print("\n--- Profils des Clusters RFM (Moyennes et Scores Inversés) ---")
print("R_Score (5=Très Récent, 1=Très Ancien) | F_Score (5=Très Fréquent, 1=Rare) | M_Score (5=Très Élevé, 1=Faible)")
print(df_cluster_analysis.to_string())

# =========================================================================
# 5. VISUALISATION (Distribution des scores R, F et M par cluster)
# =========================================================================
print("\n--- 5. Visualisation des profils de clusters ---")

df_melted_analysis = df_rfm_final.melt(
    id_vars=['Cluster_RFM'], 
    value_vars=['Recency', 'Frequency', 'Monetary'], 
    var_name='RFM_Metric', 
    value_name='Value'
)

# Crée trois boîtes à moustaches
plt.figure(figsize=(18, 5))
for i, metric in enumerate(['Recency', 'Frequency', 'Monetary']):
    plt.subplot(1, 3, i + 1)
    
    # Filtrer les outliers (les 5% les plus extrêmes) pour une meilleure visualisation
    q1 = df_melted_analysis[df_melted_analysis['RFM_Metric'] == metric]['Value'].quantile(0.05)
    q3 = df_melted_analysis[df_melted_analysis['RFM_Metric'] == metric]['Value'].quantile(0.95)
    
    sns.boxplot(
        x='Cluster_RFM', 
        y='Value', 
        data=df_melted_analysis[(df_melted_analysis['RFM_Metric'] == metric) & (df_melted_analysis['Value'] >= q1) & (df_melted_analysis['Value'] <= q3)],
        palette='viridis'
    )
    plt.title(f'Distribution de {metric} par Cluster (Outliers filtrés)', fontsize=14)
    plt.xlabel('Cluster ID', fontsize=12)
    plt.ylabel(f'{metric} (Log/Outliers Filtrés)', fontsize=12)
    plt.grid(axis='y', linestyle=':', alpha=0.6)

plt.tight_layout()
plt.show()

print("\nLe tableau d'analyse et les boîtes à moustaches montrent les profils basés sur K=3 (ou K choisi) :")
print("1. **Cluster Premium** (R élevé, F élevé, M élevé)")
print("2. **Cluster à risque** (R faible, F faible, M variable)")
print("3. **Cluster Fidèle** (F élevé, R élevé, M faible/moyen)")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# --- VÉRIFICATION ET CHARGEMENT DES DONNÉES ---
FILE_NAME = "global_superstore2.xlsx" 
# Colonnes requises : ID client, Sub-Category, Sales
features_produit = ['Customer ID', 'Sub-Category', 'Sales']

print("--- 0. Préparation des données (Chargement et nettoyage) ---")

try:
    # Chargement du fichier Excel
    df = pd.read_excel(FILE_NAME)
    
    # Nettoyage : s'assurer qu'il n'y a pas de valeurs manquantes critiques
    df_clean = df.dropna(subset=features_produit).copy()
    
    # Filtrer les transactions où les ventes sont positives
    df_clean = df_clean[df_clean['Sales'] > 0]
    
    print(f"Données chargées. Nombre de transactions : {len(df_clean)}")

except Exception as e:
    print(f"ERREUR FATALE LORS DU CHARGEMENT/PRÉ-TRAITEMENT : {e}")
    raise

# =========================================================================
# 1. PRÉPARATION DES DONNÉES DE PRÉFÉRENCE (Matrice Client-Catégorie)
# =========================================================================
print("\n--- 1. Préparation des données de préférence par catégorie ---")

# 1.1 Calcul des dépenses totales par client et par sous-catégorie
df_spending = df_clean.groupby(['Customer ID', 'Sub-Category'])['Sales'].sum().reset_index()

# 1.2 Pivotement : transformer en matrice où les colonnes sont les sous-catégories
# Les valeurs sont les montants totaux dépensés par chaque client
df_pivot = df_spending.pivot_table(
    index='Customer ID', 
    columns='Sub-Category', 
    values='Sales', 
    fill_value=0 # Remplacer les NaN (pas d'achat dans cette catégorie) par 0
)

# 1.3 Normalisation de ligne (Proportion) : Calculer la proportion de dépense
# Cela permet d'obtenir le pourcentage de dépenses de chaque client dans chaque catégorie.
# On divise chaque cellule par la somme totale des ventes de ce client.
total_sales_per_customer = df_pivot.sum(axis=1)
df_proportion = df_pivot.div(total_sales_per_customer, axis=0)

print(f"Matrice de préférence client-catégorie créée. Nombre de clients : {len(df_proportion)}")


# =========================================================================
# 2. STANDARDISATION POUR K-MEANS
# =========================================================================
print("\n--- 2. Standardisation des proportions ---")

# Normalisation : bien que les données soient déjà en proportions (0-1), 
# la standardisation (centrage et réduction) est cruciale pour K-Means
scaler_prod = StandardScaler()
df_scaled_values = scaler_prod.fit_transform(df_proportion)
df_scaled = pd.DataFrame(df_scaled_values, columns=df_proportion.columns, index=df_proportion.index)

print(f"Données standardisées. Nombre de colonnes (sous-catégories) : {df_scaled.shape[1]}")


# =========================================================================
# 3. DÉTERMINATION DU NOMBRE OPTIMAL DE CLUSTERS (Méthode du coude)
# =========================================================================
print("\n--- 3. Détermination du nombre K optimal (Méthode du coude) ---")

inertia = []
K_range = range(1, 11) 

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

# Estimation Manuelle du Coude (K_optimal à ajuster après avoir vu le graphique)
k_optimal = 3 # Fixons K=5 comme point de départ pour l'analyse de préférence

# Tracé de la méthode du coude
plt.figure(figsize=(10, 6))
plt.plot(K_range, inertia, marker='o', color='forestgreen')
plt.axvline(x=k_optimal, color='red', linestyle='--', label=f'K suggéré (À Ajuster) = {k_optimal}')
if k_optimal > 0 and k_optimal <= len(inertia):
    plt.scatter(k_optimal, inertia[k_optimal-1], color='red', s=100, zorder=5)

plt.title('Méthode du Coude (Elbow Method) pour la Segmentation par Catégorie', fontsize=15)
plt.xlabel('Nombre de clusters (K)', fontsize=12)
plt.ylabel('Inertie', fontsize=12)
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()

print(f"K fixé à {k_optimal} pour la modélisation. Veuillez ajuster K_FINAL manuellement si besoin.")

# =========================================================================
# 4. APPLICATION DE K-MEANS ET INTERPRÉTATION DES CLUSTERS
# =========================================================================
K_FINAL = k_optimal 

print(f"\n--- 4. Application de K-Means avec K={K_FINAL} ---")

kmeans_final = KMeans(n_clusters=K_FINAL, random_state=42, n_init=10)
df_proportion['Cluster_Prod'] = kmeans_final.fit_predict(df_scaled)

# 4.1. Analyse des Clusters (Moyennes des proportions)
# Les centres de clusters nous indiquent la sous-catégorie préférée de chaque groupe
df_cluster_analysis = df_proportion.groupby('Cluster_Prod').mean()
df_cluster_analysis['Nb_Clients'] = df_proportion['Cluster_Prod'].value_counts()
df_cluster_analysis = df_cluster_analysis.sort_values(by='Nb_Clients', ascending=False)


print("\n--- Profils des Clusters de Préférence (Moyennes des proportions de dépenses) ---")
print("Interprétation : Une proportion élevée (proche de 1) pour une catégorie indique une forte préférence.")

# Afficher les 3 catégories les plus importantes pour chaque cluster
print("\n--- 4.2. Synthèse des 3 principales préférences par Cluster ---")
cluster_profiles = []
for cluster in df_cluster_analysis.index:
    # Trier les proportions (hors colonne 'Nb_Clients') par ordre décroissant
    top_3 = df_cluster_analysis.loc[cluster].drop('Nb_Clients').sort_values(ascending=False).head(3)
    
    # Formater les résultats
    profile_summary = f"Cluster {cluster} ({df_cluster_analysis.loc[cluster, 'Nb_Clients']} clients):\n"
    for cat, prop in top_3.items():
        profile_summary += f"  - {cat}: {prop:.2%} des dépenses\n"
    cluster_profiles.append(profile_summary)

print("\n" + "="*50)
for profile in cluster_profiles:
    print(profile)
    print("-"*50)


# =========================================================================
# 5. VISUALISATION (Heatmap des profils)
# =========================================================================
print("\n--- 5. Visualisation des profils de clusters (Heatmap) ---")

# Exclure la colonne 'Nb_Clients' pour le graphique
df_heatmap_data = df_cluster_analysis.drop(columns=['Nb_Clients'])

plt.figure(figsize=(15, 8))
sns.heatmap(
    df_heatmap_data, 
    annot=True, 
    fmt=".2f", # Afficher 2 décimales (la proportion)
    cmap="YlGnBu", 
    linewidths=.5, 
    cbar_kws={'label': 'Proportion Moyenne des Dépenses'}
)
plt.title(f'Profils de Préférence de Catégorie par Cluster (K={K_FINAL})', fontsize=16)
plt.ylabel('Cluster ID', fontsize=12)
plt.xlabel('Sous-Catégorie', fontsize=12)
plt.yticks(rotation=0)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\nInterprétation Heatmap : Les cases de couleur foncée indiquent que ce cluster dépense une grande partie de son budget dans cette catégorie. Chaque ligne est un profil client unique.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- VÉRIFICATION ET CHARGEMENT DES DONNÉES ---
FILE_NAME = "global_superstore2.xlsx" 
# Colonnes requises : Date de commande, Sub-Category, Sales
features_reco = ['Order Date', 'Sub-Category', 'Sales']

print("--- 0. Préparation des données (Chargement et nettoyage) ---")

try:
    df = pd.read_excel(FILE_NAME)
    df_clean = df.dropna(subset=features_reco).copy()
    
    # Conversion de la date
    df_clean['Order Date'] = pd.to_datetime(df_clean['Order Date'])
    print(f"Données chargées. Nombre de transactions : {len(df_clean)}")

except Exception as e:
    print(f"ERREUR FATALE LORS DU CHARGEMENT/PRÉ-TRAITEMENT : {e}")
    raise

# =========================================================================
# 1. RECOMMANDATION SAISONNIÈRE (Tendances Temporelles)
# =========================================================================
print("\n--- 1. Analyse des Tendances Saisonnieres par Mois ---")

# 1.1 Extraction du Mois et Agrégation des Ventes
df_clean['Month'] = df_clean['Order Date'].dt.month
# Calculer la somme des ventes pour chaque combinaison (Mois, Sous-Catégorie)
df_monthly_sales = df_clean.groupby(['Month', 'Sub-Category'])['Sales'].sum().reset_index()

# 1.2 Calculer les ventes totales par mois pour normalisation
# La normalisation permet de comparer l'importance relative d'une catégorie
# même si le mois de décembre a des ventes totales plus élevées que janvier.
monthly_total_sales = df_monthly_sales.groupby('Month')['Sales'].sum()
df_monthly_sales['Sales_Proportion'] = df_monthly_sales.apply(
    lambda row: row['Sales'] / monthly_total_sales[row['Month']], 
    axis=1
)

# 1.3 Identifier les 3 catégories les plus importantes par mois
print("\n--- Top 3 des Sous-Catégories en Proportion de Ventes par Mois ---")

seasonal_recommendations = {}
for month in range(1, 13):
    df_month = df_monthly_sales[df_monthly_sales['Month'] == month]
    # Trier par proportion des ventes pour trouver le Top 3
    top_categories = df_month.sort_values(by='Sales_Proportion', ascending=False).head(3)
    
    # Formater pour l'affichage
    top_summary = []
    for _, row in top_categories.iterrows():
        top_summary.append(f"{row['Sub-Category']} ({row['Sales_Proportion']:.1%})")
        
    seasonal_recommendations[month] = ", ".join(top_summary)

# Créer un DataFrame pour une présentation claire
df_seasonal = pd.DataFrame(seasonal_recommendations.items(), columns=['Mois (1=Janv.)', 'Top 3 Catégories (Proportion)'])
print(df_seasonal.to_string(index=False))

# =========================================================================
# 2. VISUALISATION (Heatmap des tendances)
# =========================================================================
print("\n--- 2. Visualisation des tendances de vente (Heatmap) ---")

# Pivotement pour préparer les données pour la Heatmap
df_heatmap_data = df_monthly_sales.pivot_table(
    index='Sub-Category', 
    columns='Month', 
    values='Sales_Proportion', 
    fill_value=0
)

plt.figure(figsize=(15, 10))
sns.heatmap(
    df_heatmap_data, 
    annot=False, # Pour une meilleure lisibilité visuelle
    cmap="YlOrRd", # Une palette de couleurs chaudes (Jaune à Rouge)
    linewidths=.5, 
    linecolor='white',
    cbar_kws={'label': 'Proportion des Ventes Mensuelles Totales'}
)
plt.title('Tendances Saisonnieres : Proportion des Ventes par Sous-Catégorie et par Mois', fontsize=16)
plt.ylabel('Sous-Catégorie', fontsize=12)
plt.xlabel('Mois (1 = Janvier, 12 = Décembre)', fontsize=12)
plt.tight_layout()
plt.show()

print("\nLa Heatmap permet de visualiser rapidement quand une sous-catégorie (ligne) est la plus dominante (couleur foncée) au cours de l'année (colonne).")
```

```python
import pandas as pd
from itertools import combinations
from collections import Counter

# --- VÉRIFICATION ET CHARGEMENT DES DONNÉES ---
FILE_NAME = "global_superstore2.xlsx" 
# Colonnes requises : ID client, Sub-Category (Product ID pour une analyse plus fine), Order Date
features_reco = ['Customer ID', 'Order Date', 'Sub-Category']

print("--- 0. Préparation des données (Chargement et nettoyage) ---")

try:
    df = pd.read_excel(FILE_NAME)
    df_clean = df.dropna(subset=features_reco).copy()
    
    # Conversion de la date (utile si on doit recréer l'ID du panier)
    df_clean['Order Date'] = pd.to_datetime(df_clean['Order Date'])
    print(f"Données chargées. Nombre de transactions : {len(df_clean)}")

except Exception as e:
    print(f"ERREUR FATALE LORS DU CHARGEMENT/PRÉ-TRAITEMENT : {e}")
    raise

# =========================================================================
# 1. PRÉPARATION DES DONNÉES DE PANIER (Basket Creation)
# =========================================================================
print("\n--- 1. Préparation des paniers d'achat (Basket ID) ---")

# Définir ce qui constitue un "panier" d'achat. 
# Idéalement, on utilise 'Order ID'. S'il manque, on crée un ID unique par Client/Date.
if 'Order ID' not in df_clean.columns:
    print("ATTENTION : 'Order ID' non trouvé. Création d'un ID de panier unique (Client ID + Date).")
    df_clean['Basket ID'] = df_clean['Customer ID'].astype(str) + "_" + df_clean['Order Date'].dt.strftime('%Y%m%d')
    BASKET_ID_COL = 'Basket ID'
else:
    BASKET_ID_COL = 'Order ID'
    df_clean['Basket ID'] = df_clean[BASKET_ID_COL]

# Regrouper pour obtenir la liste des sous-catégories uniques dans chaque panier
df_baskets = df_clean.groupby('Basket ID')['Sub-Category'].apply(
    # On utilise set pour s'assurer que si un client achète 5 chaises, la catégorie 'Chairs' ne compte qu'une fois
    lambda x: list(set(x)) 
).reset_index(name='Items')

print(f"Nombre total de paniers uniques analysés : {len(df_baskets)}")

# =========================================================================
# 2. ANALYSE DE CO-OCCURRENCE (Association Rule Mining Simplifiée)
# =========================================================================
print("\n--- 2. Analyse de co-occurrence des paires de sous-catégories ---")

# Compteur pour stocker la fréquence de co-achat de chaque paire
co_occurrence_counts = Counter()

# Itérer sur tous les paniers
for items in df_baskets['Items']:
    # S'assurer que le panier contient au moins deux articles pour former une paire
    if len(items) >= 2:
        # Générer toutes les paires possibles (combinaisons de 2) d'articles dans ce panier
        for item_a, item_b in combinations(sorted(items), 2):
            # Le tri assure que ('Chair', 'Table') est compté de la même manière que ('Table', 'Chair')
            co_occurrence_counts[(item_a, item_b)] += 1

# =========================================================================
# 3. AFFICHAGE DES RÈGLES D'ASSOCIATION
# =========================================================================
print("\n--- 3. Top 10 des Paires de Co-Achats les plus Fréquentes ---")
print("Interprétation : Si le client achète le Produit 1, recommander le Produit 2.")

# Conversion du compteur en DataFrame pour un affichage structuré
df_co_achats = pd.DataFrame(co_occurrence_counts.items(), columns=['Paires', 'Nb_CoAchats'])

# Séparer la colonne 'Paires' en deux colonnes 'Produit 1' et 'Produit 2'
df_co_achats[['Produit 1', 'Produit 2']] = pd.DataFrame(df_co_achats['Paires'].tolist(), index=df_co_achats.index)

# Trier par le nombre de co-achats
df_co_achats = df_co_achats.sort_values(by='Nb_CoAchats', ascending=False).drop(columns=['Paires']).reset_index(drop=True)

# Afficher les résultats du top 10
print(df_co_achats.head(10).to_string(index=False))

print("\n--- Stratégie de Recommandation ---")
print("Les paires les plus fréquentes indiquent une forte complémentarité ou une habitude d'achat.")
print("Exemple : Si 'Phones' et 'Binders' sont dans le Top 3, quand un client regarde un téléphone, vous devriez lui suggérer des classeurs (et vice-versa).")
```
