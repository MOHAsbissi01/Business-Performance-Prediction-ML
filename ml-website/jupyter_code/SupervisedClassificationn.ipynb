```python
df_clean = df.copy()

print("Dataset original :", df_clean.shape)

# 1. Drop colonnes inutiles
df_clean.drop(columns=['Row ID', 'Order ID', 'Customer Name', 'Product ID', 'Postal Code'],
              inplace=True, errors='ignore')

# 2. Conversion dates + gestion immédiate des dates manquantes
df_clean['Order Date'] = pd.to_datetime(df_clean['Order Date'], dayfirst=True, errors='coerce')
df_clean['Ship Date']  = pd.to_datetime(df_clean['Ship Date'],  dayfirst=True, errors='coerce')

# Calcul du délai
df_clean['Shipping_Delay_Days'] = (df_clean['Ship Date'] - df_clean['Order Date']).dt.days
df_clean['Shipping_Delay_Days'] = df_clean['Shipping_Delay_Days'].fillna(0).clip(lower=0)

# === NOUVELLE PARTIE : on remplit les features dates AVANT de supprimer Order Date ===
df_clean['Order_Year']      = df_clean['Order Date'].dt.year.fillna(0).astype(int)
df_clean['Order_Month']     = df_clean['Order Date'].dt.month.fillna(0).astype(int)
df_clean['Order_DayOfWeek'] = df_clean['Order Date'].dt.dayofweek.fillna(0).astype(int)
df_clean['Order_Quarter']   = df_clean['Order Date'].dt.quarter.fillna(0).astype(int)

# Maintenant on peut supprimer les colonnes dates brutes en toute sécurité
df_clean.drop(columns=['Order Date', 'Ship Date'], inplace=True)

# 3. Imputation du reste
num_cols = ['Sales', 'Quantity', 'Discount', 'Profit', 'Shipping Cost', 'Shipping_Delay_Days']
df_clean[num_cols] = df_clean[num_cols].fillna(df_clean[num_cols].median())

cat_cols = df_clean.select_dtypes(include='object').columns
for col in cat_cols:
    mode_val = df_clean[col].mode()
    df_clean[col] = df_clean[col].fillna(mode_val[0] if not mode_val.empty else 'Unknown')

# 4. Flags valeurs extrêmes
p99_sales = df_clean['Sales'].quantile(0.99)
p99_ship  = df_clean['Shipping Cost'].quantile(0.99)
p01_profit = df_clean['Profit'].quantile(0.01)

df_clean['is_big_sale']       = (df_clean['Sales'] > p99_sales).astype(int)
df_clean['is_big_shipping']   = (df_clean['Shipping Cost'] > p99_ship).astype(int)
df_clean['is_big_loss']       = (df_clean['Profit'] < p01_profit).astype(int)
df_clean['is_high_discount']  = (df_clean['Discount'] >= 0.3).astype(int)

# 5. Résultat FINAL
print("\nNETTOYAGE TERMINÉ – 0 NaN GARANTI")
print("Shape final :", df_clean.shape)
print("NaN restants :", df_clean.isnull().sum().sum())   # ← 0 cette fois !
print("Lignes conservées : 51 290")

df_clean.to_csv('Global_Superstore_100%_PROPRE_51290.csv', index=False)
print("Fichier 100% propre sauvé → Global_Superstore_100%_PROPRE_51290.csv")
df_clean.head(10)
print("Colonnes numériques présentes :")
print(df_clean.select_dtypes(include='number').columns.tolist())

print("\nColonnes catégorielles présentes :")
print(df_clean.select_dtypes(include='object').columns.tolist())

print(f"\nShape final : {df_clean.shape}")
print(f"NaN restants : {df_clean.isnull().sum().sum()}")
pd.set_option('display.max_columns', None)
df_clean.head()
```

```python
# ===============================================
# PIPELINE COMPLET - PROFIT_CLASS
# ===============================================

# ==============================
# 1️⃣ Import des librairies
# ==============================
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

# ==============================
# 2️⃣ Création de la variable qualitative 'Profit_Class'
# ==============================
# On transforme la variable continue 'Profit' en 3 classes équilibrées : Low / Medium / High
df_clean['Profit_Class'] = pd.qcut(df_clean['Profit'], q=3, labels=['Low','Medium','High'])

# Vérification de la distribution
print("Distribution de Profit_Class :")
print(df_clean['Profit_Class'].value_counts())

# ==============================
# 3️⃣ Définir X et y
# ==============================
y = df_clean['Profit_Class']    # variable cible qualitative
X = df_clean.drop(columns=['Profit_Class'])

# Séparation train/test stratifiée
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
print("Distribution y_train:\n", y_train.value_counts())

# ==============================
# 4️⃣ Encoder la cible pour XGBoost
# ==============================
le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc  = le.transform(y_test)

# ==============================
# 5️⃣ Colonnes numériques et catégorielles
# ==============================
num_cols = ['Quantity','Discount','Shipping Cost','Shipping_Delay_Days',
            'Order_Year','Order_Month','Order_DayOfWeek','Order_Quarter']

cat_cols = ['Region','Sub-Category','Segment','Ship Mode']

# ==============================
# 6️⃣ Prétraitement : StandardScaler + OneHot
# ==============================
preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), num_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), cat_cols)
])

# Fit uniquement sur le train
X_train_prep = preprocessor.fit_transform(X_train)
X_test_prep  = preprocessor.transform(X_test)
print("Nombre de features après prétraitement :", X_train_prep.shape[1])

# ==============================
# 7️⃣ Sélection des features
# ==============================
selector = SelectFromModel(RandomForestClassifier(n_estimators=200, random_state=42), threshold='median')
selector.fit(X_train_prep, y_train_enc)  # utiliser labels encodés pour XGBoost
X_train_sel = selector.transform(X_train_prep)
X_test_sel  = selector.transform(X_test_prep)
print("Features avant :", X_train_prep.shape[1], "après sélection :", X_train_sel.shape[1])

# ==============================
# 8️⃣ Entraînement des modèles
# ==============================
# Random Forest
rf = RandomForestClassifier(n_estimators=300, random_state=42)
rf.fit(X_train_sel, y_train)  # labels textuels acceptés
print("Random Forest entraîné !")

# XGBoost
xgb = XGBClassifier(
    n_estimators=300,
    max_depth=5,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric='mlogloss',
    use_label_encoder=False,
    random_state=42
)
xgb.fit(X_train_sel, y_train_enc)  # labels encodés
print("XGBoost entraîné !")

# ==============================
# 9️⃣ Évaluation des modèles
# ==============================
models = {'RandomForest': rf, 'XGBoost': xgb}

for name, model in models.items():
    if name == 'XGBoost':
        y_pred_enc = model.predict(X_test_sel)
        y_pred = le.inverse_transform(y_pred_enc)
    else:
        y_pred = model.predict(X_test_sel)

    print(f"\n=== {name} ===")
    print("Accuracy :", accuracy_score(y_test, y_pred))
    print("Classification Report :\n", classification_report(y_test, y_pred))

    # Matrice de confusion graphique
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f"Confusion Matrix - {name}")
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

# ==============================
# 🔟 Sélection du meilleur modèle selon accuracy
# ==============================
best_model_name = max(models, key=lambda n: accuracy_score(y_test,
                                                          le.inverse_transform(models[n].predict(X_test_sel))
                                                          if n=='XGBoost' else models[n].predict(X_test_sel)))
best_model = models[best_model_name]
print(f"Le meilleur modèle pour prédire Profit_Class est : {best_model_name}")

# ==============================
# 1️⃣1️⃣ Sauvegarde du pipeline complet
# ==============================
joblib.dump({
    'model': best_model,
    'preprocessor': preprocessor,
    'selector': selector,
    'label_encoder': le
}, "best_model_profit_class.pkl")
print("Pipeline complet sauvegardé → best_model_profit_class.pkl")
```

```python
# ===============================================
# IMPORTS
# ===============================================
from catboost import CatBoostClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report, confusion_matrix, average_precision_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# ===============================================
# 0️⃣ Encoder les labels pour CatBoost
# ===============================================
y_train_enc = le.fit_transform(y_train)
y_test_enc  = le.transform(y_test)

# ===============================================
# 1️⃣ Définir l'espace des hyperparamètres réduit
# ===============================================
param_dist = {
    'iterations': [200, 400],  # moins d'itérations pour accélérer
    'depth': [4, 6],           # profondeur réduite
    'learning_rate': [0.03, 0.05],
    'l2_leaf_reg': [3, 5],
    'border_count': [32, 50]
}

# ===============================================
# 2️⃣ Initialiser le modèle CatBoost
# ===============================================
cat = CatBoostClassifier(
    loss_function='MultiClass',
    verbose=0,
    random_state=42,
    thread_count=-1  # utilise tous les coeurs CPU
)

# ===============================================
# 3️⃣ RandomizedSearchCV réduit
# ===============================================
random_search = RandomizedSearchCV(
    estimator=cat,
    param_distributions=param_dist,
    n_iter=10,                # moins de combinaisons pour accélérer
    scoring='balanced_accuracy',
    cv=3,
    verbose=1,
    random_state=42,
    n_jobs=-1
)

# ===============================================
# 4️⃣ Entraînement sur les données originales (sans SMOTE)
# ===============================================
random_search.fit(X_train_sel, y_train_enc)

print("Meilleurs hyperparamètres :", random_search.best_params_)

# Meilleur modèle optimisé
best_cat = random_search.best_estimator_

# ===============================================
# 5️⃣ Prédictions et évaluation
# ===============================================
y_pred_enc = best_cat.predict(X_test_sel)
y_pred = le.inverse_transform(y_pred_enc)
y_proba = best_cat.predict_proba(X_test_sel)

# PR AUC multi-classes
pr_auc_cat = average_precision_score(
    np.eye(len(le.classes_))[y_test_enc],  # one-hot des labels test
    y_proba,
    average='macro'
)

print("\n=== CatBoost Classifier (Optimisé, sans SMOTE) ===\n")
print("Classification Report:\n", classification_report(y_test, y_pred))
print("PR AUC (macro) :", round(pr_auc_cat, 3))

# ===============================================
# 6️⃣ Matrice de confusion graphique
# ===============================================
cm_cat = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm_cat, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title("Matrice de confusion - CatBoost Optimisé")
plt.xlabel("Prédit")
plt.ylabel("Réel")
plt.show()

# ===============================================
# 7️⃣ Importance des features
# ===============================================
importances = best_cat.get_feature_importance()
feature_names = X_train_sel.shape[1]*[f"Feature_{i}" ]  # si pas de nom exact après sélection

indices = np.argsort(importances)[::-1]
sorted_importances = importances[indices]
sorted_features = np.array(feature_names)[indices]

plt.figure(figsize=(10, 6))
sns.barplot(x=sorted_importances, y=sorted_features, palette="viridis")
plt.title("Feature Importances - CatBoost Optimisé")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.show()

# ===============================================
# 8️⃣ Ajouter CatBoost aux résultats
# ===============================================
results = {}
results['CatBoost'] = pr_auc_cat
```

```python
# ===============================================
# 🔵 SVM - PIPELINE COMPLET - PROFIT_CLASS
# ===============================================
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, average_precision_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# ==============================
# 1️⃣ Sous-échantillon pour RandomizedSearch rapide
# ==============================
X_sample, _, y_sample, _ = train_test_split(
    X_train_sel, y_train,  # utiliser X_train_sel issu de la sélection de features
    train_size=15000,
    stratify=y_train,
    random_state=42
)

# ==============================
# 2️⃣ Définition de l'espace des hyperparamètres
# ==============================
param_dist = {
    'C': [0.1, 1, 10, 50, 100],
    'gamma': ['scale', 0.01, 0.1, 0.5],
    'kernel': ['rbf']
}

# ==============================
# 3️⃣ RandomizedSearchCV sur l'échantillon
# ==============================
svm = SVC(class_weight='balanced', probability=True, random_state=42)

rand_svm = RandomizedSearchCV(
    estimator=svm,
    param_distributions=param_dist,
    n_iter=15,
    scoring='accuracy',  # pour pipeline Profit_Class, on se base sur accuracy
    cv=3,
    n_jobs=-1,
    random_state=42,
    verbose=1
)
rand_svm.fit(X_sample, y_sample)
print("Meilleurs paramètres SVM :", rand_svm.best_params_)

# ==============================
# 4️⃣ Réentraînement sur tout le train set
# ==============================
best_svm = SVC(
    C=rand_svm.best_params_['C'],
    gamma=rand_svm.best_params_['gamma'],
    kernel='rbf',
    class_weight='balanced',
    probability=True,
    random_state=42
)
best_svm.fit(X_train_sel, y_train)
print("SVM entraîné !")

# ==============================
# 5️⃣ Prédictions et évaluation
# ==============================
y_pred_svm = best_svm.predict(X_test_sel)

print("\n=== SVM ===")
print("Accuracy :", accuracy_score(y_test, y_pred_svm))
print("Classification Report :\n", classification_report(y_test, y_pred_svm))

# Matrice de confusion graphique
cm_svm = confusion_matrix(y_test, y_pred_svm)
plt.figure(figsize=(6,5))
sns.heatmap(cm_svm, annot=True, fmt='d', cmap='Greens', cbar=False)
plt.title("Confusion Matrix - SVM")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# ==============================
# 6️⃣ Optionnel : PR AUC multi-classes
# ==============================
from sklearn.preprocessing import LabelBinarizer
lb = LabelBinarizer()
y_test_bin = lb.fit_transform(y_test)
y_proba_svm = best_svm.predict_proba(X_test_sel)
pr_auc_svm = average_precision_score(y_test_bin, y_proba_svm, average='macro')
print("PR AUC (macro) SVM :", round(pr_auc_svm, 3))

# ==============================
# 7️⃣ Ajouter SVM au dictionnaire des modèles
# ==============================
models['SVM'] = best_svm
results['SVM'] = pr_auc_svm
```
